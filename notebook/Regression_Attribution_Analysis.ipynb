```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("/Users/eric/ecommerce_sql_project/data/regression_data.csv")

df.head()
```

```python
#特征工程
df['log_order_value'] = np.log1p(df['order_value'])

df.describe()
```

```python
#建立回归模型
model = smf.ols(
    'log_order_value ~ frequency + product_count + recency',
    data=df
).fit(cov_type='HC3')   # 使用稳健标准误

print(model.summary())
```

回归归因结论:

1.商品种类（product_count）是订单金额的最强驱动因素（β = 0.018） 

2.最近活跃客户订单金额更高（β = -0.0011）

3.在控制商品种类和活跃度后，频次对单笔金额影响较小且略为负向
